# Clase 02 - Derivadas discretas (diferencias finitas)

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Aproximación hacia adelante

Usemos la siguiente aproximación de primer orden para la derivada
$$ f'(x) = \frac{f(x+h)-f(x)}{h}$$
donde $f(x)=\cos(x)$

```python
def funcion(x):
    return (np.cos(x))
```

```python
def df(x,h):
    return ((funcion(x+h)-funcion(x))/h)
```

```python
def fp_exacta(x):
    return (-np.sin(x))
```

```python
h = 0.1
x = np.arange(0,2*np.pi,h)
```

```python
fp = df(x,h)
plt.plot(x,fp,label='aproximación numérica')
plt.plot(x,fp_exacta(x),'--',label='valor exacto')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend();
```

```python
plt.plot(x,fp-fp_exacta(x))
plt.xlabel('$x$')
plt.ylabel('$f\'_{num}-f\'_{exacta}$');
```

### Aproximación centrada

Ahora, usemos la aproximación de segundo orden para la derivada
$$ f'(x) = \frac{f(x+h)-f(x-h)}{2h}$$

```python
def df2(x,h):
    return ((funcion(x+h)-funcion(x-h))/(2*h))
```

```python
fp = df2(x,h)
plt.plot(x,fp,label='aproximación numérica')
plt.plot(x,fp_exacta(x),'--',label='valor exacto')
plt.xlabel('$x$')
plt.ylabel('$f(x)$')
plt.legend();
```

```python
plt.plot(x,fp-fp_exacta(x))
plt.xlabel('$x$')
plt.ylabel('$f\'_{num}-f\'_{exacta}$');
```

```python
paso = []
error_max = []

h = 1.0
max = 1.0 
while max > 1.0e-6:
    paso.append(h)
    x = np.arange(0,2.0*np.pi,h)
    y = funcion(x)
    fp = df(y,h)
    max = np.max(np.abs(fp-fp_exacta(y)))
    error_max.append(max)
    h /= 2
```

```python
paso2 = []
error_max2 = []

h = 1.0
max = 1.0 
while max > 1.0e-10:
    paso2.append(h)
    x = np.arange(0,2.0*np.pi,h)
    y = funcion(x)
    fp = df2(y,h)
    max = np.max(np.abs(fp-fp_exacta(y)))
    error_max2.append(max)
    h /= 2

plt.loglog(paso,error_max,'o',label='forward difference')
plt.loglog(paso2,error_max2,'ro',label='central difference')
plt.xlabel('$h$')
plt.ylabel('error maximo')
plt.legend();
```
