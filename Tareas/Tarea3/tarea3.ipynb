## P1


```python
import numpy as np
from scipy.interpolate import CubicSpline
import matplotlib.pyplot as plt

def f(x):
    return np.sin(np.pi * x) + 0.5 * x**2

x_points = np.array([0, 1, 2, 3, 4])
y_points = f(x_points)

spline_cubic = CubicSpline(x_points, y_points, bc_type='natural')


x_eval = np.array([1.5, 2.5])
y_eval_natural = spline_cubic(x_eval)


y_real = f(x_eval)


error_natural = np.abs(y_eval_natural - y_real)
print("Errores con condiciones de borde naturales:", error_natural)


x_fine = np.linspace(0, 4, 100)
plt.plot(x_fine, spline_cubic(x_fine), label='Spline Cúbico (Natural)')
plt.scatter(x_eval, y_eval_natural, color='red', label='Estimaciones')
plt.plot(x_fine, f(x_fine), label='Función Exacta', linestyle='--')
plt.scatter(x_points, y_points, color='blue', label='Puntos dados')
plt.legend()
plt.title('Spline Cúbico con Condiciones de Borde Naturales')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python

pendiente1 = 1 
pendiente2 = -1   

spline_custom = CubicSpline(x_points, y_points, bc_type=((1, pendiente1), (1, pendiente2)))


y_eval_custom = spline_custom(x_eval)


error = np.abs(y_eval_custom - y_real)
print("Error:", error)
plt.plot(x_fine, spline_custom(x_fine), label='Spline Cúbico (Pendientes dadas)')
plt.scatter(x_eval, y_eval_custom, color='red', label='Estimaciones')
plt.plot(x_fine, f(x_fine), label='Función Exacta', linestyle='--')
plt.scatter(x_points, y_points, color='blue', label='Puntos dados')
plt.legend()
plt.title('Spline Cúbico con Pendientes dadas')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

```python
from scipy.integrate import quad

def f(x):
    return np.exp(-x**2)


points = np.array([-np.sqrt(3/5), 0, np.sqrt(3/5)])
weights = np.array([5/9, 8/9, 5/9])


integral_gauss = np.sum(weights * f(points))
print("Integral usando Gauss-Legendre (n=3):", integral_gauss)


integral_quad, _ = quad(f, -1, 1)
print("Integral usando scipy quad:", integral_quad)
```
